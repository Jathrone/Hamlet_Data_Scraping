```python
def extract_speaker(response):
    return response.content.decode("utf-8").split("<br/>\n")[1].split("#")[1].split("_")[0]
def extract_listeners(response):
    return response.content.decode("utf-8").split("</h2>")[1].split("<br/>")[:-1]
```

Here I am trying to massage the speakers and listeners list so that both lists are equal. This is assuming that everyone who ever appears onstage would have a speaking part. However, it appears that PLAYERS doesn't have a corresponding speaker label. This could be due to PLAYER being the collection of all players. This doesn't illuminate the difference between PLAYERS.0.1 and PLAYERS.1, however. To better get around this, I will instead assign speakers and groups line by line. 

```python
listener_groups = []
def register_group(listeners):
    listeners = [all_listeners[k] for k in listeners]
    listeners.sort()
    for i, group in enumerate(listener_groups):
        if listeners == group:
            return i
    else:
        listener_groups.append(listeners)
        return len(listener_groups) - 1
```

```python
whole_text_line_index = 1
lines_dict = {}
for i in range(2, 4168):
    four_letter_num = ("0000" + str(i))[-4:]
    four_letter_num_minus = ("0000" + str(i-1))[-4:]
    try:
        listener_group = extract_listeners(listeners_dict[four_letter_num_minus])
        listener_group_id = register_group(listener_group)

        speaker = all_speakers[extract_speaker(speakers_dict[four_letter_num])]
        whole_text_line_index += 1
    except:
        if four_letter_num == "3075":
            whole_text_line_index += 1
        print(four_letter_num)
    if speaker not in listener_groups[listener_group_id]:
        continue
    else:
        lines_dict[four_letter_num] = {"listener_group_id": listener_group_id, "speaker":speaker, "line":whole_text_lines[whole_text_line_index]}
#         break
lines_dict['0001'] = {'listener_group_id': 0,
                     'speaker': 'Barnardo',
                     'line': "Who's there?"}
```
